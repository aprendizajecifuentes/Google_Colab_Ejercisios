Análisis de Ventas por Categoría en una Tienda

```python
import pandas as pd
url_ventas="https://drive.google.com/uc?export=download&id=1JS7jkpbWMaOlpsh_7Q6P8nue1U9aiM-C"
df_ventas=pd.read_csv(url_ventas)
df_ventas.head()

## Agrupamos los datos por la columna categoria

df_ventas_agrupado=df_ventas.groupby("Categoría").sum("ventas")
df_ventas_agrupado.head()

```

###**Ejercisio 2 ,Sesion 8, Und 8:**
  Cálculo del Precio Unitario Promedio por Categoría

```python
## Agrupamos los datos por la columna categoria añadir columna precio promedio

df_ventas_agrupado=df_ventas.groupby("Categoría").mean("Precio Unitario").round(2)

df_ventas_agrupado.head()
```

###**Ejercisio 3 ,Sesion 8, Und 8:**
  Cálculo del Precio Unitario Promedio por Categoría

```python
## Agrupamos los datos por la columna categoria añadir columna precio promedio

df_ventas_prodcuto=df_ventas.groupby("Producto").sum("ventas")
display(df_ventas_prodcuto[['Cantidad Vendida', 'Precio Unitario']])

```
