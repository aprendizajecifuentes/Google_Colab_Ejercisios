# **Conversión de Columnas Numéricas con Manejo de Errores**

Se convertira los numero string a numero y los que generen error y no se conviertan sera remplazados por **NaN**



Objetivo del ejercicio:

- Cargar los datos desde el archivo datos_conversion.csv en un DataFrame de pandas.
- Convertir las columnas "Ventas" y "Precio Unitario" a valores numéricos utilizando pd.to_numeric().
- Reemplazar los valores no convertibles (errores) con NaN.
- Mostrar las primeras 10 filas del DataFrame resultante para verificar los cambios.

```python
#1. Cargar los datos desde el archivo datos_conversion.csv en un DataFrame de pandas.
import pandas as pd

df = pd.read_csv('datos_conversion.csv')
print(df)
```

```python
#2. Convertir las columnas "Ventas" y "Precio Unitario" a valores numéricos utilizando pd.to_numeric().
#Reemplazar los valores no convertibles (errores) con NaN.
df['Ventas'] = pd.to_numeric(df['Ventas'], errors='coerce')
df['Precio Unitario'] = pd.to_numeric(df['Precio Unitario'], errors='coerce')
print(df)
```

```python
#4. Mostrar las primeras 10 filas del DataFrame resultante para verificar los cambios.
print(df.head(10))
```
