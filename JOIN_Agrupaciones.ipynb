###**AGRUPACION DE TABLAS JOIN_(INNER, RIGTH,LEFTH)**
  Fusión de Datos con Inner Join

```python

import pandas as pd

# Cargar los datos desde los archivos CSV
ventas_df = pd.read_csv('https://drive.google.com/uc?export=download&id=1txHNdJmFQqfGr1p7_9Y07HjfaHV7oJ1b')
productos_df = pd.read_csv('https://drive.google.com/uc?export=download&id=1Vbfvur9orXO3lbb9RQgu27hAg1F9BMKm')

# Realizar la fusión (inner join) en base a la columna "Producto_ID"
merged_df = pd.merge(ventas_df, productos_df, on='Producto_ID', how='inner')

# Mostrar las primeras 10 filas del DataFrame resultante
display(merged_df.head(10))
```

###**Ejercisio 2 ,Sesion 7, Und 8:**

 Fusión de Datos con left Join


```python
# Realizar la fusión (left join) en base a la columna "Producto_ID"
merged_df = pd.merge(ventas_df, productos_df, on='Producto_ID', how='left')
display(merged_df.head(10))
```

###**Ejercisio 3 ,Sesion 7, Und 8:**

 Fusión de Datos con rigth Join

```python
# Realizar la fusión (inner join) en base a la columna "Producto_ID"
merged_df = pd.merge(ventas_df, productos_df, on='Producto_ID', how='right')
display(merged_df.head(10))
```

```python
# Buscar filas con valores nulos en el DataFrame merged_df
filas_con_nulos = merged_df[merged_df.isnull().any(axis=1)]

# Mostrar las filas encontradas
display(filas_con_nulos)
```

###**Ejercisio 4 ,Sesion 7, Und 8:**

 Fusión de Datos con rigth Join

```python
# Establecer 'Producto_ID' como índice en ambos DataFrames
ventas_indexed_df = ventas_df.set_index('Producto_ID')
productos_indexed_df = productos_df.set_index('Producto_ID')

# Realizar la fusión utilizando el método .join()
joined_df = ventas_indexed_df.join(productos_indexed_df, how='inner')

# Mostrar las primeras 10 filas del DataFrame resultante
display(joined_df.head(10))
```

###**Ejercisio 5 ,Sesion 7, Und 8:**

 Cálculo de "Ingresos Totales" tras la Fusión de Datos


```python
# Establecer 'Producto_ID' como índice en ambos DataFrames
ventas_indexed_df = ventas_df.set_index('Producto_ID')
productos_indexed_df = productos_df.set_index('Producto_ID')


# Realizar la fusión utilizando el método .join()
joined_df = ventas_indexed_df.join(productos_indexed_df, how='inner')

# Calcular el ingreso total por cada producto
joined_df['Ingreso_Total'] = joined_df['Cantidad Vendida'] * joined_df['Precio Unitario']


# Mostrar las primeras 10 filas del DataFrame resultante
display(joined_df.head(10))
```
