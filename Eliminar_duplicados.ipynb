# **ELIMINA DUPLICADOS**
Elimina Segun productos duplicadas o Fechas que sean iguales(Duplicadas)

**UNIDAD 8 SESION 5 EJERCISIO 1 (DUPLICADOS)**


```python
import pandas as pd

url="https://drive.google.com/uc?export=download&id=1d8ccNG7I55lMGurBiZy7pQNk78DRkQDV"
df=pd.read_csv(url)


print("\n num de filas: ",len(df))

print("\n Conteo valores nulos: ", df.isnull().sum())

#elimino filas con valores nulos
df_limpio=df.dropna()

print("filas limpias de valores nulos: ",len(df_limpio))

#se verifica que todo este limpio y reluciente
print("\n ¿aun quedan valores nulos?", df_limpio.isnull().any().any())

#identificar filas duplicadas
num_duplicadas = df_limpio.duplicated().sum()
print("num filas duplicadas: ",num_duplicadas)

#mostrar filas duplicadas
filas_duplicadas = df_limpio[df_limpio.duplicated()]
print(filas_duplicadas.sort_values(by=['Producto']))


```

**UNIDAD 8 SESION 5
EJERCISIO 2 (ELIMINAR DUPLICADOS)**

```python
#Eliminar filas duplicadas
df_limpio_2v=df_limpio.drop_duplicates()
print(df_limpio_2v.sort_values(by=['Producto']))

#numero de filas duplicadas borradas
print("filas borradas: ",len(df_limpio_2v)-len(df_limpio))
```

**UNIDAD 8 SESION 5
EJERCISIO 3 (ELIMINAR DUPLICADOS SEGUN VENTAS Y PRODUCTOS)**

```python
#IDENTIFICAR DATOS DUPLICADOS SEGUN PRODUCTO Y VENTAS
duplicados_prod_ventas = df_limpio[df.duplicated(subset=['Producto', 'Ventas'], keep=False)]

print(duplicados_prod_ventas.sort_values(by=['Producto']))

#cantidad duplicados por producto y ventas
print("num filas duplicadas",len(duplicados_prod_ventas))


```

```python
# eliminar filas duplicadas conservando precio unitario mayor
df_filtrado = df_limpio.sort_values(by="Precio Unitario", ascending=False) \
                .drop_duplicates(subset=["Producto", "Ventas"], keep="first")

print(df_filtrado.sort_values(by="Precio Unitario", ascending=False))

#Mostrar el número total de filas antes y después de la eliminación
print("\nfilas antes de eliminacion:",len(df_limpio))
print("\nfilas despues de eliminacion:",len(df_filtrado))
print("\nfilas eliminadasion:",len(df_filtrado)-len(df_limpio))

```

**UNIDAD 8 SESION 5 EJERCISIO 4 (ELIMINAR DUPLICADOS SEGUN FECHA)**

```python
# filas duplicdas segun fecha
duplicadas_fecha = df_limpio[df.duplicated(subset='Fecha', keep=False)]
print(duplicadas_fecha.sort_values(by="Fecha").head(8))

#Eliminar solo los duplicados más recientes (última aparición),
df_sin_duplicados_fecha = df[~df.duplicated(subset='Fecha', keep='first')]
print("\nDespues de eliminar duplicados por fecha\n")
print(df_sin_duplicados_fecha.sort_values(by="Fecha").head(8))

#Mostrar el número total de filas antes y después.
print("\nfilas antes de eliminacion:",len(df_limpio))
print("\nfilas despues de eliminacion:",len(df_sin_duplicados_fecha))
print("\nfilas eliminadasion:",len(df_sin_duplicados_fecha)-len(df_limpio))
```
