# **IDENTIFICACION VALORES NULOS** (Identifica, enlista y borra valores nulos)

**Ejercisio 1 UNIDAD 8 SESION 4** (Identificación de Valores Nulos)

```python
#abrir archivo excel csv de nuestro computador
import pandas as pd

url="https://docs.google.com/spreadsheets/d/18LQO7KKhV8kQ_rxegxFDAcscKxs5Drii/export?format=csv"
df=pd.read_csv(url)

print("\n num de filas: ",len(df))

print("\n Conteo valores nulos: ", df.isnull().sum())

#elimino filas con valores nulos
df_limpio=df.dropna()

print("filas limpias de valores nulos: ",len(df_limpio))

#se verifica que todo este limpio y reluciente
print("\n ¿aun quedan valores nulos?", df_limpio.isnull().any().any())
```
