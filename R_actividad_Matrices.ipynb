```r
## Creamos la matriz 3x3
notas <- matrix(1:9, nrow = 3, byrow = TRUE)
print("Creamos la matriz 3x3")
print(notas)
cat("\n")

## obtenemos el valor de fila 2, columna 3
print("obtenemos el valor de fila 2, columna 3")
f2_c3<-notas[2,3]
print(f2_c3)
cat("\n")

# Todos los elementos de la columna 1
print("Todos los elementos de la columna 1")
C1<-notas[,1]
print(C1)
cat("\n")

#Todos los elementos de la fila 3
print("Todos los elementos de la fila 3")
f3<-notas[3,]
print(f3)
cat("\n")

#La suma de la primera fila.
print("La suma de la primera fila.")
sum_f1<-sum(notas[1,])
print(sum_f1)
cat("\n")

#La transpuesta de la matriz.
print("La transpuesta de la matriz.")
T_matrix<-t(notas)
print(T_matrix)
cat("\n")

#Crea otra matriz llamada extra con los números del 10 al 18 (3x3)
# combínala con notas usando:cbind(notas, extra)

extra<- matrix(10:18,nrow=3,byrow = TRUE)
Combinada<- cbind(notas, extra)
print("Matrix combinada")
print(Combinada)



```
