```r
#Dato numerico
x<- 3.14
print(x)
print(class(x)) #Se genera el numerico

#Dato entero
y<- 10L
print(y)
print(class(y))

#Dato logico
z<- TRUE
print(z)
print(class(z))

#Dato caracter
w<- "Hola Mundo"
print(w)
print(class(w))

#Dato complejo o complex
v<- 1+2i
print(v)
print(class(v))

#Factor
Colores<- factor(c("Gris","Naranja","Amarillo","Gris"))
print(Colores)
print(class(Colores))

#Tipo de dato de cada variable
sapply(list(x,y,z,w,v,Colores),class)
```

**ACTIVIDAD DE CLASE**

Crea cinco variables con diferentes tipos de datos:

- numero <- 15.8           # numérico
- entero <- 10L            # entero
- logico <- TRUE           # lógico
- texto <- "R lenguaje"    # carácter
- complejo <- 2 + 3i       # complejo

Agrupa las variables en una lista llamada mis_datos:

* mis_datos <- list(numero, entero, logico, texto, complejo)


* Aplica sapply() para conocer más sobre cada variable:

Tipo de dato:

* sapply(mis_datos, class)

```r
#Dato numerico
P<- 15.8
print(P)
print(class(P))

#Dato entero
T<- 10L
print(T)
print(class(T))

#Dato logico
B<- TRUE
print(B)
print(class(B))

#Dato caracter
H<- "R lenguaje"
print(H)
print(class(H))

#Dato complejo o complex
L<- 2+3i
print(L)
print(class(L))

mis_datos <- list(numeric, integer, logical, character, complex)
sapply(mis_datos, class)
```
