```python
import pandas as pd

url="https://drive.google.com/uc?export=download&id=1Oeo3UjU3Si9i7rumO7_G0bY68suoiwJk"
df=pd.read_csv(url)

print("********* Datois guardados ********")
print(df.head(5))

```

```python
df["Ventas"] =pd.to_numeric(df["Ventas"],errors="coerce")
df["Precio Unitario"]=pd.to_numeric(df["Precio Unitario"],errors="coerce")

print("**** tipos de datos *****")
print(df.dtypes)
```

```python

from dateutil import parser

# Función robusta para convertir fechas
def convertir_fecha(fecha):
    try:
        # Convertimos a string, limpiamos espacios, y parseamos
        return parser.parse(str(fecha).strip(), dayfirst=True)
    except:
        return pd.NaT

# Aplicar la función a la columna 'Fecha' y guardar en nueva columna
df['Fecha'] = df['Fecha'].apply(convertir_fecha)

# Formatear como 'YYYY-MM-DD' (opcional)
df['Fecha'] = df['Fecha'].dt.strftime('%Y-%m-%d')

# Mostrar resultados
print(df.head(10))

```
